## Velocidad de la nave 
Muestra los segundos necesarios para pasar de 0 a 11200 metros por segundo, dada la acelaraci√≥n de la nave en metros por segundo.

```python
endVelocity = 11200
startVelocity = 0 
accelaration = 9.8 

time = (endVelocity - startVelocity) / accelaration
print("Tiempo para alcanzar la velocidad deseada = ", time)
```
